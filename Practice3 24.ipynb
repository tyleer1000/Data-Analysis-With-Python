```python
'''
Return only Acuras with Horsepower over 250
'''
#data.loc[(data['Make'].isin("Acura")), 'Horsepower'] > 250
data_Make = (data['Make'].isin(["Acura"]))
data_HP = (data['Horsepower'] > 250)
data[data_HP & data_Make]

data["Case"] = 'bad'
data.loc[data_HP & data_Make, "Case"] = 'good'
data

```

```python
def timestwo(val):
    return val * 2
data2 = data.copy()
data2[['EngineSize', 'Cylinders']] = data[['EngineSize', 'Cylinders']].map(timestwo)
data2
```

```python
#Get the average invoice by make

data.groupby(['Origin','Make']).agg({'Horsepower': 'mean'})

dataclean = data.copy()

dataclean['Invoice'] =dataclean['Invoice'].str.replace('$', "")
dataclean['Invoice'] =dataclean['Invoice'].str.replace(',', "")

dataclean['Invoice'] = dataclean['Invoice'].astype(float)
dataclean['Invoice'] = dataclean['Invoice'].round()
dataclean.groupby(['Make']).agg({('Invoice'): 'mean'})
```
