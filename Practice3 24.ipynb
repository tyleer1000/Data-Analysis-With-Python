```python
'''
Return only Acuras with Horsepower over 250
'''
#data.loc[(data['Make'].isin("Acura")), 'Horsepower'] > 250
data_Make = (data['Make'].isin(["Acura"]))
data_HP = (data['Horsepower'] > 250)
data[data_HP & data_Make]

data["Case"] = 'bad'
data.loc[data_HP & data_Make, "Case"] = 'good'
data

```

```python
def timestwo(val):
    return val * 2
data2 = data.copy()
data2[['EngineSize', 'Cylinders']] = data[['EngineSize', 'Cylinders']].map(timestwo)
data2
```

```python
#Get the average invoice by make

data.groupby(['Origin','Make']).agg({'Horsepower': 'mean'})

dataclean = data.copy()

dataclean['Invoice'] =dataclean['Invoice'].str.replace('$', "")
dataclean['Invoice'] =dataclean['Invoice'].str.replace(',', "")

dataclean['Invoice'] = dataclean['Invoice'].astype(float)
dataclean.groupby(['Make']).agg({('Invoice'): 'mean'}).round(2)
```

```python
df_student.groupby(['gender']).agg({'mark': 'mean'}).round(2)
```

```python
'''
What is the percentage of students with marks greater than or equal to 60? 
First, find total # of students with marks >= 60
Then, compare that # to the total # of students to calc. percentage'''

df_marks = df_student.copy()
is_high_marks = df_marks['mark'] >= 60
df_marks[is_high_marks]

high_marks_student = df_marks[is_high_marks]
str(round(high_marks_student.shape[0]/df_marks.shape[0]*100, 2)) + '%'
```
